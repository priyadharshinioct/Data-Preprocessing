#### Data Cleaning

```python
env_data=env_data.drop(columns=["Unnamed: 0","Temp C"],axis=1)
env_data
```

```python
del env_data["Ozone"]
```

```python
env_data
```

```python
env_data[env_data["Solar.R"].isna()]
```

```python
env_data.groupby("Weather")["Solar.R"].mean().round()
```

```python
env_data["Solar.R"]=env_data.groupby("Weather")["Solar.R"].transform(lambda x:x.fillna(x.mean().round()))
env_data
```
